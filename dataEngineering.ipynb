###### Normalizing the whole numerical values in the dataset

```python
def scaling(column):

    min_value = column.min()
    max_value = column.max()
    scaled_column = (column - min_value) / (max_value - min_value)
    return scaled_column

# Apply Min-Max scaling function to each column
normalized_df = df_numerical.apply(scaling)
```

```python
normalized_df
```

###### now we normalize the dataset


```python
std_list = normalized_df.std()
```

```python
std_list
```

```python
df_std=pd.DataFrame(std_list, columns=['standard deviation'])
```

```python
df_std
```

```python
threshhold=0.3
standard_deviation_exceeds=df_std[df_std['standard deviation'] > threshhold]
```

```python
standard_deviation_exceeds_columns=standard_deviation_exceeds['standard deviation']
```
